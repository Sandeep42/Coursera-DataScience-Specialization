```r
df <- read.table("featuresextracted.csv", header= TRUE, stringsAsFactors = FALSE, sep=",")
```

```r
head(df)
```

```r
df$sex <- NULL
```

```r
df$activity[df$activity=="liedown_bed"] <- "bed_liedown_getup"
df$activity[df$activity=="getup_bed"] <- "bed_liedown_getup"
df$activity[df$activity=="eat_meat"] <- "eat"
df$activity[df$activity=="eat_soup"] <- "eat"
df$activity[df$activity=="standup_chair"] <- "chair_standup_sitdown"
df$activity[df$activity=="sitdown_chair"] <- "chair_standup_sitdown"
df$activity[df$activity=="climb_stairs"] <- "stairs_climb_descend"
df$activity[df$activity=="descend_stairs"] <- "stairs_climb_descend"
```

```r
df$activity <- as.factor(df$activity)
```

```r
str(df)
```

```r
require(caret)
```

## Create test, train and validation sets

```r
set.seed(998)
inTraining <- createDataPartition(df$activity, p = .80, list = FALSE)
```

```r
training <- df[inTraining,]
validation  <- df[-inTraining,]
dim(training)
dim(validation)
```

```r
in1 <- createDataPartition(training$activity, p = .60, list = FALSE)
```

```r
training2 <- training[in1,]
testing  <- training[-in1,]
dim(training2)
dim(testing)
```

## Create models

### RF


```r
fitControl <- trainControl(## 10-fold CV
                           method = "repeatedcv",
                           number = 10,
                           ## repeated ten times
                           repeats = 10)
```

```r
set.seed(825)
rfFit1 <- train(activity ~ ., data = training2,
                 method = "rf",
                 trControl = fitControl)
```

```r
rfpredict <- predict(rfFit1, testing)
```

```r
rfconfusion <- confusionMatrix(rfpredict, testing$activity)
```

```r
rfconfusion$overall
```

```r
rfdfconfusion <- as.data.frame(rfconfusion$table)
dim(rfdfconfusion)
```

```r
plot <- ggplot(rfdfconfusion)
plot + geom_tile(aes(x=Reference, y=Prediction, fill=Freq)) + scale_x_discrete(name="Actual Class") + scale_y_discrete(name="Predicted Class") + scale_fill_gradient(low="red",high="white")+
 theme(axis.text.x = element_text(angle = 45, hjust = 1))+ labs(fill="Normalized\nFrequency")
```

```r
plot(rfFit1)
```

```r
rfconfusion$overall
```

### GBM

```r
set.seed(825)
gbmFit1 <- train(activity ~ ., data = training2,
                 method = "gbm",
                 trControl = fitControl,
                 ## This last option is actually one
                 ## for gbm() that passes through
                 verbose = FALSE)
```

```r
gbmpredict <- predict(gbmFit1, testing)
```

```r
gbmconfusion <- confusionMatrix(gbmpredict, testing$activity)
```

```r
gbmconfusion$overall
```

```r
gbmdfconfusion <- as.data.frame(gbmconfusion$table)
dim(gbmdfconfusion)
```

```r
plot <- ggplot(gbmdfconfusion)
plot + geom_tile(aes(x=Reference, y=Prediction, fill=Freq)) + scale_x_discrete(name="Actual Class") + scale_y_discrete(name="Predicted Class") + scale_fill_gradient(low="red",high="white")+
 theme(axis.text.x = element_text(angle = 45, hjust = 1))+ labs(fill="Normalized\nFrequency")
```

### LDA

```r
set.seed(825)
ldaFit1 <- train(activity ~ ., data = training2,
                 method = "lda",
                 trControl = fitControl,
                 ## This last option is actually one
                 ## for gbm() that passes through
                 verbose = FALSE)
```

```r
ldapredict <- predict(ldaFit1, testing)
```

```r
ldaconfusion <- confusionMatrix(ldapredict, testing$activity)
```

```r
ldaconfusion$overall
```

```r
ldadfconfusion <- as.data.frame(ldaconfusion$table)
dim(rfdfconfusion)
```

```r
plot <- ggplot(ldadfconfusion)
plot + geom_tile(aes(x=Reference, y=Prediction, fill=Freq)) + scale_x_discrete(name="Actual Class") + scale_y_discrete(name="Predicted Class") + scale_fill_gradient(low="red",high="white")+
 theme(axis.text.x = element_text(angle = 45, hjust = 1))+ labs(fill="Normalized\nFrequency")
```

## Creating an ensemble model

```r
ensemData <- data.frame(rf= rfpredict,
                        gbm= gbmpredict,
                        lda= ldapredict, actual= testing$activity 
                            )
```

```r
ensemFit <- train(actual~., data= ensemData, method="LogitBoost")
```

```r
ensemPred <- predict(ensemFit, ensemData)
```

```r
enconfusion <- confusionMatrix(ensemPred, testing$activity)
```

```r
enconfusion$overall
```
